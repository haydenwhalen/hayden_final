```python
import json
import requests
import pandas as pd

# Define the dates you want to collect data for
dates = ['2023-08-01', '2023-08-02', '2023-08-03', '2023-08-04', '2023-08-05']

# Your API key (ensure this is kept private)
key = "cOAx9g8QtgcOnZKKI0U_gp1h81NHPej3"
headers = {"Authorization": f"Bearer {key}"}

# Initialize the DataFrame to store the results
df = pd.DataFrame(columns=[
    'date', 'symbol', 'open', 'high', 'low', 'close', 'volume', 'afterHours', 'preMarket'
])

# Loop through each date and fetch the data
for date in dates:
    url = f"https://api.polygon.io/v1/open-close/AAPL/{date}?adjusted=true"

    # Make the API request
    req = requests.get(url, headers=headers)

    # Parse the response as JSON
    result = json.loads(req.text)

    try:
        # Store the response data in the DataFrame
        df.loc[len(df)] = [
            result['from'],
            result['symbol'],
            result['open'],
            result['high'],
            result['low'],
            result['close'],
            result['volume'],
            result['afterHours'],
            result['preMarket']
        ]
    except KeyError:
        print(f"Error with result data for {date}: {result}")

# Display the DataFrame
print(df)








```
